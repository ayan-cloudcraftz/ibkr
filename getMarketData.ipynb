```python
from ibapi.client import *
from ibapi.wrapper import *
import pandas as pd
```

```python
def getOptionPrice(contract):

    class OptionPrice(EClient, EWrapper):
        def __init__(self):
            EClient.__init__(self, self)
            self.BID = None
            self.ASK = None
            self.LOW = None
            self.HIGH = None
            self.OPEN = None
            self.CLOSE = None
            self.VOLUME = None
            self.BID_SIZE = None
            self.ASK_SIZE = None
            self.LAST_SIZE = None

        # getiing valid id to obtain market data
        def nextValidId(self, orderId: int):
            mycontract = contract
            self.reqMarketDataType(4)
            self.reqMktData(orderId, mycontract, "101", 0, 0, [])

        # Called on market data changes
        def tickPrice(self, reqId, tickType, price, attrib):
            print(f"tickPrice. reqId: {reqId}, tickType: {TickTypeEnum.to_str(tickType)}, price: {price}, attribs: {attrib}")
            if TickTypeEnum.to_str(tickType) == 'DELAYED_BID' and price > 0:
                self.BID = price
            elif TickTypeEnum.to_str(tickType) == 'DELAYED_ASK' and price > 0:
                self.ASK = price
            elif TickTypeEnum.to_str(tickType) == 'DELAYED_HIGH':
                self.HIGH = price
            elif TickTypeEnum.to_str(tickType) == 'DELAYED_LOW':
                self.LOW = price
            elif TickTypeEnum.to_str(tickType) == 'DELAYED_OPEN':
                self.OPEN = price
            elif TickTypeEnum.to_str(tickType) == 'DELAYED_CLOSE':
                self.CLOSE = price

            if (self.BID != None and self.ASK != None and self.HIGH!= None and self.LOW!= None and self.OPEN!= None and self.CLOSE!= None and self.VOLUME!= None and self.BID_SIZE!= None and self.ASK_SIZE!= None and self.LAST_SIZE!= None):
                self.disconnect()

        # Called on market data changes
        def tickSize(self, reqId, tickType, size):
            print(f"tickSize. reqId: {reqId}, tickType: {TickTypeEnum.to_str(tickType)}, size: {size}")
            if TickTypeEnum.to_str(tickType) == 'DELAYED_VOLUME':
                self.VOLUME = size
            elif TickTypeEnum.to_str(tickType) == 'DELAYED_BID_SIZE':
                self.BID_SIZE = size
            elif TickTypeEnum.to_str(tickType) == 'DELAYED_ASK_SIZE':
                self.ASK_SIZE = size
            elif TickTypeEnum.to_str(tickType) == 'DELAYED_LAST_SIZE':
                self.LAST_SIZE = size

            if (self.BID != None and self.ASK != None and self.HIGH!= None and self.LOW!= None and self.OPEN!= None and self.CLOSE!= None and self.VOLUME!= None and self.BID_SIZE!= None and self.ASK_SIZE!= None and self.LAST_SIZE!= None):
                self.disconnect()

            
            

    app = OptionPrice()
    app.connect("127.0.0.1", 7497, 1100)    #building connection for getting market data
    app.run()

    return app
```
